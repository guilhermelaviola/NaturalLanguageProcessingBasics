## **Example: Sentiment Analysis**
The following example uses the TextBlob library to analyze the sentiment of a sentence. The model processes the text, assigns a polarity score, and classifies the sentiment as positive, negative, or neutralâ€”illustrating how NLP techniques can be applied to interpret human language computationally.

```python
# Example text:
text = 'I really love how easy this app is to use!'

# Creating a TextBlob object:
blob = TextBlob(text)

# Getting sentiment polarity (-1 = negative, +1 = positive):
sentiment = blob.sentiment.polarity

# Displaying the result:
if sentiment > 0:
    print('Positive sentiment ðŸ˜Š')
elif sentiment < 0:
    print('Negative sentiment ðŸ˜ž')
else:
    print('Neutral sentiment ðŸ˜')

print(f'Sentiment score: {sentiment}')
```
